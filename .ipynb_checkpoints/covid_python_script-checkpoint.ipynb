```python
import pandas as pd
data = pd.read_csv('C:/Users/1push/Downloads/Covid dataset/covid_data.csv')
```

```python
import matplotlib.pyplot as mt
```

```python
dt = data.groupby('region')['deaths'].sum().reset_index().sort_values(by = 'deaths',ascending = False).head(10)
mt.figure(figsize = (16,5))
mt.bar(dt['region'],dt['deaths'],label = 'Covid Top 10 States of Deaths 2020-22',width = 0.5,edgecolor = 'black',color = 'salmon')
mt.grid(axis = 'y')
mt.xlabel('States',fontsize = 12,color = 'blue',fontweight = 'bold')
mt.ylabel('Deaths',fontsize = 12,fontweight = 'bold')
mt.title('Top 10 States of Covid Deaths',fontweight = 'bold')
mt.xticks(rotation = 45,fontsize = 10)
mt.legend(loc = 'best')
mt.tight_layout()
mt.savefig('covid_top_10_deaths.png', dpi=300, bbox_inches='tight')
mt.show()
```

```python
#2020 covid cases
data['date'] = pd.to_datetime(data['date'])

year_2020 = data[data['date'].dt.year == 2020].copy()
year_2020['month_2020'] = year_2020['date'].dt.month

year_2021 = data[data['date'].dt.year == 2021].copy()
year_2021['month_2021'] = year_2021['date'].dt.month

year_2022 = data[data['date'].dt.year == 2022].copy()
year_2022['month_2022'] = year_2022['date'].dt.month







cvd_month20 = year_2020.groupby('month_2020')['confirmed'].sum().reset_index()
cvd_month21 = year_2021.groupby('month_2021')['confirmed'].sum().reset_index()
cvd_month22 = year_2022.groupby('month_2022')['confirmed'].sum().reset_index()



mt.figure(figsize = (10,5),facecolor = 'lightgreen')

mt.title('Monthly Covid Case: 2020, 2021, 2022')
mt.gca().set_facecolor('white')
mt.xlabel('Month Wise Cases',color = 'blue',fontweight = 'bold')
mt.ylabel('Covid Spread',fontweight = 'bold')
mt.xticks(ticks= range(1,13),labels =['January','February','March','April','May','June','July','August','September','October','November','December'],rotation = 45,fontsize = 10)




mt.plot(cvd_month20['month_2020'],cvd_month20['confirmed'],label = 'Y-20 Month Covid Cases',color = 'blue',marker = 'o')
mt.plot(cvd_month21['month_2021'],cvd_month21['confirmed'],label = 'Y-21 Month Covid Cases',color = 'red',marker = '*')
mt.plot(cvd_month22['month_2022'],cvd_month22['confirmed'],label = 'Y-22 Month Covid Cases',color = 'green',marker = '.')










mt.legend(loc = 'best')
mt.tight_layout()
mt.savefig('covid_monthlyreports.png', dpi=300, bbox_inches='tight')

mt.show()
```

```python
#year wise data show as in PIE CHART:-
explode = [0.1,0,0]
yearly_cvd_cases = [year_2020['confirmed'].sum(),year_2021['confirmed'].sum(),year_2022['confirmed'].sum()]

mt.pie(yearly_cvd_cases,labels = ['2020','2021','2022'],autopct = '%0.1f%%',explode = explode)

mt.title('Covid Cases of 2020-2021-2022',fontweight = 'bold')
mt.tight_layout()
mt.savefig('Yearly_covid_cases_piechart.png', dpi=100, bbox_inches='tight')
mt.show()
```
